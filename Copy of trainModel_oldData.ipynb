```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'
#from google.colab import drive
#drive.mount('/content/drive')

# Fetch preprocessed data
data = 'preprocessed.csv'
fe_data = pd.read_csv(data, usecols=["book_title", "book_desc", "bow"])
fe_data.head()

fe_data.shape[0] - fe_data.nunique()


fe_data1 = fe_data[:1000]
```

```python
#!pip install keybert
from keybert import KeyBERT
kw_model = KeyBERT()

def get_keywords(text):
    keywords = kw_model.extract_keywords(text, keyphrase_ngram_range=(1, 1), stop_words="english")
    keywords = " ".join([k[0] for k in keywords])
    return keywords
```

```python
fe_data1["keywords"] = fe_data1.book_desc.apply(get_keywords)
```

```python
fe_data1.keywords.head()
```

```python
fe_data1["keywords"] = fe_data1[['bow', 'keywords']].fillna('').agg(' '.join, axis=1)
fe_data1.drop(['bow', 'book_desc'], axis = 1, inplace=True)
```

```python
fe_data1[fe_data1.duplicated(subset=["book_title"], keep="first")]
```

```python
fe_data1 = fe_data1.drop_duplicates(subset=["book_title"], keep='first')
```
